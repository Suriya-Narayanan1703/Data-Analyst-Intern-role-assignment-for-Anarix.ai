### Finding numerical column values distribution using histogram plot 

```python
# Plot histograms for numerical columns
numerical_columns = ['Sales', 'Quantity', 'Discount', 'Profit', 'Shipping Cost']
plt.figure(figsize=(15, 10))
for i, column in enumerate(numerical_columns, 1):
    plt.subplot(2, 3, i)
    sns.histplot(df[column].dropna(), kde=True, bins=30)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

### Visualizing with bar plot histogram and pie chart

```python
def visualize(column):
    # Set up the figure with 3 subplots
    fig, axes = plt.subplots(1, 3, figsize=(18, 6))
    
    # Bar Plot
    sns.barplot(x=column.index, y=column.values, ax=axes[0], palette='viridis')
    axes[0].set_title('Bar Plot')
    axes[0].set_xlabel('Value')
    axes[0].set_ylabel('Count')
    
    # Histogram
    sns.histplot(x=column.index, weights=column.values, ax=axes[1], color='blue')
    axes[1].set_title('Histogram')
    axes[1].set_xlabel('Value')
    axes[1].set_ylabel('Count')
    
    # Box Plot
    axes[2].pie(column.values, labels=column.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("viridis", len(column)))
    axes[2].set_title('Pie Chart')
    
    # Adjust layout
    plt.tight_layout()
    plt.show()

```

```python
visualize(frequent_values["Ship Mode"])
```

## Time series visualization

### Year wise visualization of sales

```python
df['Year'] = df['Order Date'].dt.year
sales_by_year = df.groupby('Year')['Sales','Profit'].sum().reset_index()
plt.figure(figsize=(10, 6))
plt.plot(sales_by_year['Year'], sales_by_year['Sales'], marker='o', linestyle='-', color='b')
plt.plot(sales_by_year['Year'], sales_by_year['Profit'], marker='*', linestyle='-', color='g')
plt.title('Year-wise Sales')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.grid(True)
plt.xticks(sales_by_year['Year'])  # Ensure all years are shown on the x-axis
plt.tight_layout()
plt.show()
```

### Result:
### --the sales got drastically improved on the yearly basis
### --the profit got slightly improved

### Month wise visulalization

```python
year=int(input("enter year between 2011 to 2014:")) 
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
df_year = df[df['Year'] == year]

monthly_data = df_year.groupby('Month')[['Sales', 'Profit']].sum().reset_index()
# Plot Sales
plt.plot(monthly_data['Month'], monthly_data['Sales'], marker='o', linestyle='-', color='b', label='Sales')
# Plot Profit
plt.plot(monthly_data['Month'], monthly_data['Profit'], marker='o', linestyle='-', color='g', label='Profit')

plt.title('Month-wise Sales and Profit for')
plt.xlabel('Month')
plt.ylabel('Amount')
plt.grid(True)
plt.xticks(monthly_data['Month'], ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.legend()  # Add a legend to distinguish between Sales and Profit
plt.tight_layout()
plt.show()
```

### Results:
### --- Both profit and sales has fluctuations but they are increasing
